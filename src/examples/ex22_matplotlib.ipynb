# Matplotlib으로 시각화 해보기

### 라이브러리 임포트

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

### 첫 번째 플롯(사건의 논리적인 패턴과 배치; 여기서는 그래프) 그리기

```python
x = np.linspace(0, 2 * np.pi, 200)    # 0에서 2π까지의 값을 200개의 점으로 나눈 x 배열을 생성
y = np.sin(x)                         # x 값에 대응하는 사인 함수의 값을 포함하는 배열

fig, ax = plt.subplots()              # fig: 그래프 전체, ax: 그래프 개별 축(x축, y축)
ax.plot(x, y)                         # x와 y 값에 해당하는 선형 그래프 그리기
plt.show()                            # 그래프를 화면에 출력
```

### 두 번째 플롯 그리기 (time, temp, hum)

```python
time = [11, 12, 13, 14, 15]
temp = [30.5, 29.0, 30.0, 29.5, 31.0]
hum = [70.0, 71.5, 69.0, 69.5, 70.5]

fig, ax = plt.subplots()
ax.plot(time, temp, label="Temp", color="r")   # temp 선형 그래프
ax.plot(time, hum, label="Hum", color="b")       # hum 선형 그래프
ax.set_xlabel("Time")
ax.set_ylabel("Values")
ax.set_title("Temp & hum over Time")
ax.legend()   # 범례 표시
plt.show()
```

```python
plt.style.available   # 플롯 스타일
```

### 플롯에서 산점도 확인하기 (붓꽃 분류)

```python
# 데이터셋 임포트
from sklearn.datasets import load_iris

# 붓꽃 데이터셋 로드
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# 종 이름을 추가
species_map = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

# 산점도 행렬 그리기
sns.pairplot(iris_df, hue='species')
plt.show()
```

### 플롯에서 산점도 확인하기 (time, temp, hum)

```python
time = [11, 12, 13, 14, 15]
temp = [30.5, 29.0, 30.0, 29.5, 27.0]
hum = [70.0, 71.5, 69.0, 69.5, 70.5]


plt.figure(figsize=(10, 6)) # 온도와 습도의 산점도 그리기 (10 ✕ 6 사이즈 플롯)

plt.scatter(time, temp, label='Temperature (°C)', color='r')  # 온도 산점도
plt.scatter(time, hum, label='Humidity (%)', color='b')       # 습도 산점도

plt.title('Temp & Hum over Time Scatter')
plt.xlabel('Time (hours)')
plt.ylabel('Values')
plt.xlim(0, 20)   # x축 표시 범위를 0부터 20까지로 설정
plt.legend()
plt.show()
```
