# AHT21 센서값 기반으로 데이터 시각화 해보기

### 라이브러리 임포트

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### 데이터 불러오기

```python
# CSV 파일 경로 (ex20_data_extract_rtc_aht21.py에서 추출된 ex20.csv를 가상 드라이브에 업로드)
csv_file = "./ex20.csv"

# CSV 파일 읽어서 데이터프레임으로 변환
df = pd.read_csv(csv_file, names=["Time", "Temp", "Hum"])

# 데이터프레임의 선두 5개의 데이터 확인하기
df.head()
```

### 첫번째 플롯 (시간별 온습도 변화)

```python
fig, ax = plt.subplots()
ax.plot(df["Time"], df["Temp"], label="Temperature (°C)", color="r", marker=".")   # temp 선형 그래프
ax.plot(df["Time"], df["Hum"], label="Humidity (%)", color="b", marker=".")     # hum 선형 그래프

plt.title('Temp & Hum over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Values')
plt.xticks(rotation=90)         # X축 라벨 눈금 90°회전
plt.tight_layout()
plt.legend()
plt.grid()        # 플롯 내부 격자 설정
plt.show()
```

### 두번째 플롯 (시간별 온습도 변화 산점도)

```python
fig, ax = plt.subplots()

plt.scatter(df["Time"], df["Temp"], label='Temperature (°C)', color='r', marker=".")  # 온도 산점도
plt.scatter(df["Time"], df["Hum"], label='Humidity (%)', color='b', marker=".")       # 습도 산점도

plt.title('Temp & Hum over Time Scatter')
plt.xlabel('Time (hours)')
plt.ylabel('Values')
plt.xticks(rotation=90)

ax.xaxis.set_major_locator(plt.MaxNLocator(25))     # 주요 25개 지점만 표시

plt.tight_layout()
plt.legend()
plt.grid()
plt.show()
```
